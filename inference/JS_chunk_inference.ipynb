```python
import chromadb
from chromadb.config import Settings
from FlagEmbedding import BGEM3FlagModel

# Configuration
CHROMA_DIR = "./chroma_db"
COLLECTION_NAME = "js_code_knowledge"

# Initialize ChromaDB
client = chromadb.PersistentClient(
    path=CHROMA_DIR,
    settings=Settings(anonymized_telemetry=False)
)

collection = client.get_collection(name=COLLECTION_NAME)

# Load embedding model ONCE
model = BGEM3FlagModel(
    "BAAI/bge-m3",
    use_fp16=False  # Mac-safe
)

print("✅ Setup complete. Model and ChromaDB loaded.")
```

```python
# Change your query here
query = "what are various schemecode validations"
OUTPUT_FILE = "retrieved_chunks.txt"
TOP_K = 10

# Encode query
query_embedding = model.encode(query)["dense_vecs"]

# Query ChromaDB
results = collection.query(
    query_embeddings=query_embedding.tolist(),
    n_results=TOP_K
)

docs = results["documents"][0]
distances = results["distances"][0]   # ⬅️ relevance scores
metas = results["metadatas"][0]

# Save retrieved chunks with distance
with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
    f.write(f"Query: {query}\n")
    f.write("=" * 80 + "\n\n")

    for i, (doc, dist, meta) in enumerate(zip(docs, distances, metas), 1):
        f.write(f"Result {i}\n")
        f.write(f"Distance (lower = better): {dist:.6f}\n")
        f.write(f"Type: {meta.get('chunk_type', 'N/A')}\n")
        f.write(f"File: {meta.get('file_path', 'N/A')}\n")
        f.write("-" * 40 + "\n")
        f.write(doc)
        f.write("\n\n" + "=" * 80 + "\n\n")

        print(f"\n{i}. Distance: {dist:.6f}")
        print(f"   {doc[:200]}...")

print(f"\n✅ Results with relevance scores saved to '{OUTPUT_FILE}'")
```

```python
def get_llm_response(user_query, retrieved_chunks):
    """
    Generate explanation using Groq LLM based on retrieved context
    """
    # Build context from retrieved chunks
    context = "\n\n".join([
        f"[Chunk {i+1} - {meta.get('chunk_type', 'N/A')} from {meta.get('file_path', 'N/A')}]\n{doc}"
        for i, (doc, meta) in enumerate(zip(retrieved_chunks['docs'], retrieved_chunks['metas']))
    ])

    system_prompt = """You are an expert JavaScript frontend developer with deep knowledge of jQuery, DOM manipulation, AJAX patterns, event handling, form validation, and modern web application UI/UX patterns.

CRITICAL INSTRUCTIONS - NO HALLUCINATION:
1. ONLY use information explicitly present in the provided code context
2. DO NOT make assumptions about code that is not shown
3. DO NOT invent function names, event handlers, selectors, or API endpoints
4. If information is not available in the context, clearly state "This information is not available in the provided code"
5. ONLY explain what you can directly see in the retrieved code chunks
6. DO NOT speculate about backend endpoints, server responses, or database operations unless explicitly shown in the code
7. If a complete UI workflow cannot be determined from the provided context, acknowledge the gaps

Your role is to:
- Analyze ONLY the provided JavaScript code context thoroughly
- Answer the user's query directly and naturally based STRICTLY on what is shown in the code
- Adapt your response format to match the nature of the query (detailed workflow explanation, quick function summary, code comparison, etc.)
- Highlight relevant JavaScript patterns, jQuery usage, and DOM manipulation that are EXPLICITLY visible
- Provide comprehensive explanations WITHOUT generating new code
- Be honest about limitations - if the context doesn't show something, say so

Focus on answering the user's specific question accurately and concisely based solely on the provided code. Be detailed when needed, brief when appropriate, but never invent or assume information beyond what is explicitly shown."""

    user_message = f"""USER QUERY: {user_query}

RETRIEVED CODE CONTEXT:
{context}

IMPORTANT: Answer the user's query based STRICTLY on the code provided above. Do not make assumptions or add information not present in the context.

Provide a clear, natural response that directly addresses the user's question. Adapt your answer format to match what the user is asking for - whether that's a detailed workflow explanation, a quick function summary, a list of related components, or any other format that best answers their specific query.

Remember:
1. Answer naturally based on the query type and scope
2. ONLY use information explicitly shown in the retrieved code chunks
3. Do not hallucinate or assume implementation details
4. Focus on frontend/client-side perspective
5. If the query asks for something not shown in the code, clearly state that"""

    try:
        # Call Groq API
        chat_completion = groq_client.chat.completions.create(
            messages=[
                {"role": "system", "content": system_prompt},
                {"role": "user", "content": user_message}
            ],
            model="llama-3.3-70b-versatile",  # You can change to other models like "mixtral-8x7b-32768"
            temperature=0.3,
            max_tokens=4096,
            top_p=0.9
        )

        return chat_completion.choices[0].message.content

    except Exception as e:
        return f"Error generating LLM response: {str(e)}"


# Generate LLM explanation
print("\n" + "="*80)
print("🤖 GENERATING LLM EXPLANATION...")
print("="*80 + "\n")

retrieved_data = {
    'docs': docs,
    'metas': metas,
    'distances': distances
}

llm_response = get_llm_response(query, retrieved_data)

# Save LLM response
LLM_OUTPUT_FILE = "llm_js_explanation.txt"
with open(LLM_OUTPUT_FILE, "w", encoding="utf-8") as f:
    f.write(f"Query: {query}\n")
    f.write("=" * 80 + "\n\n")
    f.write("RESPONSE:\n")
    f.write("-" * 80 + "\n\n")
    f.write(llm_response)
    f.write("\n\n" + "=" * 80 + "\n")

print(llm_response)
print(f"\n✅ JavaScript explanation saved to '{LLM_OUTPUT_FILE}'")
```
